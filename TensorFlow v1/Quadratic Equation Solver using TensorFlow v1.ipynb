```
%tensorflow_version 1.x

import tensorflow as tf
from tensorflow.python.framework import ops
```

```
a = tf.constant(float(input("Enter a: ")))
b = tf.constant(float(input("Enter b: ")))
c = tf.constant(float(input("Enter c: ")))

x = tf.Variable(0.0, trainable=True)

LR = 0.1 #Learning Rate
```

```
y = (a*x*x) + (b*x) + c
# y = ax^2 + bx + c
```

```
optAdam = tf.train.AdamOptimizer(LR).minimize(y)
#Adam Optimizer
```

```
init = tf.global_variables_initializer()
#Initialise All Variables
```

```
sess = tf.Session()
#Start the session
```

```
sess.run(init)
#Initialise All Variables
```

```
for i in range(100000):

  sess.run(optAdam)

  if(sess.run(y)<=0):
    #Computing till Y = 0
    break
```

```
print(sess.run([x]))
```
